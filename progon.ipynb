```python
t= []
h =[]

for k in range(len(df)):
    if df.iloc[k,3]=="timestamp":
        for i in range(352):
            if df.iloc[k,i] == "bpm":
                t.append(df.iloc[k,4])
                h.append(df.iloc[k,i-1])
print(t)
print(h)  
```

```python
dt = pd.DataFrame(t)
dt.rename(columns = {0 : "Timestamp"},inplace = True )
dt['Heart_rate'] = h
```
