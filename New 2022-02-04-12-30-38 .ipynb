```python
import pandas as pd
df = pd.read_csv("2022-02-04-12-30-38.csv")
df
df
```

```python
t= []
h =[]

for k in range(6112):
    if df.iloc[k,3]=="timestamp":
        for i in range(352):
            if df.iloc[k,i] == "bpm":
                t.append(df.iloc[k,4])
                h.append(df.iloc[k,i-1])
print(t)
print(h)
            
```

```python
dt = pd.DataFrame(t)
dt.rename(columns = {0 : "Timestamp"},inplace = True )
dt['Heart_rate'] = h
```

```python
dt
```

```python
import random
hand = []
ds = [int(dt.iloc[k,1]) for k in range(810)]
print(ds)
print(len(ds))
for i in ds:
    if i >= 135:
        hand.append(int(random.uniform(i-(0.15*i), i+(0.15*i))))
    else:
        hand.append(int(random.uniform(i-(0.05*i), i+(0.05*i))))
```

```python
dt['Heart_rate_hand'] = hand
dt
```

```python
#dt.to_csv("res_2022-02-04-12-30-38.csv")
```

Построение модели:

```python
dm = dt.copy(deep=True)
#dm = dm.drop(columns='Heart_rate') 
dm
```

```python
dm['Timestamp'] = dm['Timestamp'].astype('int64')
dm['Heart_rate'] = dm['Heart_rate'].astype('int64')
```

```python
X = dm[['Timestamp','Heart_rate_hand']]
y = dm['Heart_rate']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn import *
model_LR = linear_model.LinearRegression()
model_LR.fit(X_train, y_train)
model_Ridge = linear_model.Ridge()
model_Ridge.fit(X_train, y_train)
model_Lasso = linear_model.Lasso()
model_Lasso.fit(X_train, y_train)
model_SVM = svm.SVR()
model_SVM.fit(X_train, y_train)
```

```python
y_pred_LR = model_LR.predict(X_test)
y_pred_Ridge = model_Ridge.predict(X_test)
y_pred_Lasso = model_Lasso.predict(X_test)
y_pred_SVM = model_SVM.predict(X_test)
```
